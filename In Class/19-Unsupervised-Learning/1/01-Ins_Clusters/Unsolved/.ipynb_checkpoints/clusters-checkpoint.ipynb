```python
# Import the modules
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# Generate three synthetic clusters
X, y = make_blobs(
    centers=3, 
    random_state=1
)
```

```python
# Visualizing all three classes
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
# Generate five synthetic clusters
# Generate three synthetic clusters
X, y = make_blobs(
    centers=6, 
    random_state=42
    ,n_samples=500,
    cluster_std=1.0
)
```

```python
y
```

```python
colors = ["green","red","blue","yellow"]
```

```python
# Visualizing all five classes
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
# Generate synthetic clusters
```

```python
# Visualizing all classes
from sklearn.cluster import KMeans
```

```python
# create model
model = KMeans(n_clusters=6)
```

```python
model
```

```python
model.fit(X)
```

```python
model_labels = model.labels_  
model_labels
```

```python
import pandas as pd
```

```python
pd.DataFrame(y,model_labels).reset_index()
```

```python
model_p = model.predict(X)
```

```python
pd.DataFrame(y,model_p)
```
