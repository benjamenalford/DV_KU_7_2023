```python
from bs4 import BeautifulSoup
import requests
```

```python
base_url = 'https://topeka.craigslist.org/search/foa#search=1~gallery~0~6'
```

```python
response = requests.get(base_url)
```

```python
cl_soup = BeautifulSoup(response.text)
```

```python
##cl-search-result
cl_items = cl_soup.find_all('li',class_='cl-static-search-result')
```

```python
cl_items[0].find('div',class_="title").text
cl_items[0].find('div',class_="price").text
cl_items[0].find('a')['href']
```

```python
cl_item={}
cl_stuff = []
```

```python
for c in cl_items:
    cl_item={}
    cl_item["title"] = c.find('div',class_="title").text
    cl_item["price"] = c.find('div',class_="price").text
    cl_item["url"] = c.find('a')['href']
    cl_stuff.append(cl_item)
```

```python
cl_stuff
```

```python
import pandas as pd
```

```python
cl_df = pd.DataFrame(cl_stuff)
```

```python
cl_df.tail()
```

```python
cl_df.to_csv("cl.csv")
```

```python
url = 'https://en.wikipedia.org/wiki/Jim_Carrey_filmography'
```

```python
jim_film_soup = BeautifulSoup(requests.get(url).text)
```

```python
jim_table_soup = jim_film_soup.find(id="bodyContent").find('table')
```

```python
jim_rows = jim_table_soup.find('tbody').find_all('tr')
```

```python
films =[]
for r in jim_rows:
    film = {}
    tds = r.find_all('td')
    if (len(tds)>0):
        film['title'] = tds[0].text.strip()
        film['role'] = tds[1].text.strip()
        film['url'] = tds[0].find('a')['href']
#    for td in r.find_all('td'):
#        td_text = td.text.strip()
#        print(td_text)
    films.append(film)
    print("*******")
```
