```python
from bs4 import BeautifulSoup
import requests
```

```python
base_url = 'https://topeka.craigslist.org/search/foa#search=1~gallery~0~6'
```

```python
response = requests.get(base_url)
```

```python
cl_soup = BeautifulSoup(response.text)
```

```python
##cl-search-result
cl_items = cl_soup.find_all('li',class_='cl-static-search-result')
```

```python
cl_items[0].find('div',class_="title").text
cl_items[0].find('div',class_="price").text
cl_items[0].find('a')['href']
```

```python
cl_item={}
cl_stuff = []
```

```python
for c in cl_items:
    cl_item={}
    cl_item["title"] = c.find('div',class_="title").text
    cl_item["price"] = c.find('div',class_="price").text
    cl_item["url"] = c.find('a')['href']
    cl_stuff.append(cl_item)
```

```python
cl_stuff
```

```python
import pandas as pd
```

```python
cl_df = pd.DataFrame(cl_stuff)
```
