```python
from bs4 import BeautifulSoup 
import requests
```

```python
url = 'https://www.cnn.com/'
```

```python
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text)
```

```python
cnn_soup = soup.find('div', class_='stack_condensed').find_all('div',class_="container__headline")
```

```python
for headline in cnn_soup:
    print(headline.text.strip())
```
