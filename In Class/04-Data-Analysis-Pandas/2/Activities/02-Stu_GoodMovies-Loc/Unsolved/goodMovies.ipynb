```python
# Dependencies
import pandas as pd
```

```python
# Load in File from resources
# 'movie_scores.csv'
file_path = "Resources/movie_scores.csv"
```

```python
# Read and display with pandas
movies_df  = pd.read_csv(file_path)
movies_df.head(2)
```

```python
# List all the columns the table provides
cols = movies_df.columns
new_col=[]

for c in cols:
    if (c.startswith('IMDB') |c.startswith('FILM') ):
        new_col.append(c)
new_col     
```

```python
new_col = [c for c in cols ]
```

```python
# only care about Imdb, so create a new table that takes the Film and all the columns relating to IMDB
imdb_df = movies_df[new_col]
imdb_df.head(2)
```

```python
# List only movies whose ratings are over 7 (out of 10) in IMDB
good_movies_df = imdb_df.loc[imdb_df['IMDB'] >=7.0]
good_movies_df.tail(2)
```

```python
# Find lesser-known movies to watch, with fewer than 20K votes
unknown_df = good_movies_df.loc[good_movies_df['IMDB_user_vote_count'] <20000]
unknown_df.tail()
```

```python
# Finally, export this file to an Excel spreadsheet -- without the DataFrame index.
unknown_df.to_csv("MY movie list.csv",index=False)
```
