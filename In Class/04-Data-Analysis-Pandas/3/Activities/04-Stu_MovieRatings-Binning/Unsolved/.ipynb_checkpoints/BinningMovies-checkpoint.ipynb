```python
# Import Dependencies
import pandas as pd
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
csv_path = "Resources/movie_scores.csv"
movies_df = pd.read_csv(csv_path)

movies_df.head()
```

```python
# Figure out the minimum and maximum IMDB user vote count
imdb_min = movies_df['IMDB_user_vote_count'].min()
print(imdb_min)
imdb_max = movies_df['IMDB_user_vote_count'].max()
print(imdb_max)
```

```python
list(range(0,imdb_max,int(imdb_max/10.0)))
```

```python
# Create bins in which to place values based upon IMDB vote count
bins=list(range(imdb_min,imdb_max,int(imdb_max/10.0)))
# Create labels for these bins
labels=["<33659","<67075","<100491", "<133907", "<167323", "<200739", "<234155", "<267571", "<300987"]
labels
```

```python
# Slice the data and place it into bins
# Place the data series into a new column inside of the DataFrame
movies_df['User Vote Range']=pd.cut(movies_df['IMDB_user_vote_count'],bins,labels=labels)
movies_df.head(2)
```

```python
# Create a GroupBy object based upon "IMDB User Votes Group"
movies_df.groupby('User Vote Range').count()['FILM']

# Find how many rows fall into each bin


# Get the average of each of the first 5 rating columns within the GroupBy object
movies_df.groupby('User Vote Range').mean()[['IMDB','Metacritic','RottenTomatoes']]
```
