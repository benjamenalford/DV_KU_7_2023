```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
temps_f = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
temps_f
```

```python
#Assign to the x-axis a range of numerical values representing each month of the year.
months = range(1,13)
#np.arange(1,13)
months
```

```python
temps_c = []
for temp in temps_f:
    c = (temp - 32) * 0.56
    temps_c.append(c)
temps_c
```

```python
temps_c = [(temp - 32) * 0.56 for temp in temps_f ]
temps_c
```

```python
#Plot the degrees Fahrenheit points.
plt.plot(months, temps_f)
plt.xlabel("Months")
plt.ylabel("temp")
```

```python
plt.plot(months, temps_c)
plt.xlabel("Months")
plt.ylabel("temp")
```

```python
plt.plot(months, temps_c,label='Celsius', marker='*',color='blue')
plt.plot(months, temps_f,label ="Fairenough",color="red")
plt.hlines(32, 0, 12,color="red")
plt.hlines(0, 0, 12,color="blue")
plt.xlabel("Months")
plt.ylabel("temp")
plt.legend(loc='lower right')
plt.savefig("newJersey.png")
plt.show()
```
