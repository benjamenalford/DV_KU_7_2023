```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Danger_Drop= [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
Rail_Gun= [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
time = np.arange(0,130,10)
```

```python
plt.plot(time,Danger_Drop,color="red",label="Danger Drop")
plt.plot(time,Rail_Gun,color="blue",label="Rail_Gun")
plt.xlabel("Coaster Runtime")
plt.ylabel("Speed (mph)")
plt.title("Coaster Speed Over Time")
plt.legend(loc='upper right')
plt.xlim(0,120)
plt.ylim(5,95)
plt.savefig("danger_zone.png")
plt.grid()
```

```python
plt.bar(time,Danger_Drop,color="red",label="Danger Drop")
plt.grid()
```
