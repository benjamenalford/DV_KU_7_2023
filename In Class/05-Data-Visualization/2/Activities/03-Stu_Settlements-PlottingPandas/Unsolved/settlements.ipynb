```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read CSV
path = "Resources/union_settlements_1995.csv"
df = pd.read_csv(path)
df.head(2)
```

```python
# Get total settlements by union
settlements_df = df['UNION'].value_counts()
settlements_df
```

```python
# Configure plot, figsize, title, and axis labels
fig = settlements_df.plot(kind="bar",facecolor="red",figsize=(8,6), title="Major Collective Bargining Settlements")
plt.xlabel("Union")
plt.ylabel("Settlements")
# Configure x-tick rotation
xticklabels = settlements_df.index
fig.set_xticklabels(xticklabels,rotation=45,rotation_mode="anchor",ha="right",wrap=True)
plt.tight_layout()
# Show plot
plt.show()
```
