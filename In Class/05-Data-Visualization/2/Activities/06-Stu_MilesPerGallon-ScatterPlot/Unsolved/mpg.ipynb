```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
path = "../Resources/mpg.csv"
df = pd.read_csv(path)
df.head()
```

```python
car_df = df.set_index('car name')
car_df[['mpg','horsepower']].plot(kind="bar")
```
