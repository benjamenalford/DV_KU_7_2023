```python
# Dependencies
import json
import requests 
```

```python
# Specify the URL
result_count = 50
min_wows =1
url = f'https://owen-wilson-wow-api.onrender.com/wows/random?results={result_count}&wow_in_movie={min_wows}'
# Make request and store response
response = requests.get(url)
#200 all good
#300 redirected / moved
#400 - you probably made a mistake - 404 not found , 403 0not authorised
#500 - the api is broken
response
```

```python
# JSON-ify response
response = response.json()
```

```python
for wow in response:
    print( wow['movie'],wow['full_line'],wow['total_wows_in_movie'])
```
